```python
import tkinter as tk
from tkinter import filedialog
from tkinter import font as tkfont

# 파일 열기 함수 (가져오기 기능)
def 가져오기():
    file_path = filedialog.askopenfilename(title="파일을 선택하세요", filetypes=(("텍스트 파일", "*.txt"), ("모든 파일", "*.*")))
    if file_path:
        with open(file_path, 'r', encoding='utf-8') as file:
            data = file.read()
            print("파일 내용:")
            print(data)

# tkinter 기본 설정
root = tk.Tk()
root.title("영단어 관리 프로그램")  # 창 제목 설정
root.geometry("1500x1000")  # 창 크기 설정

# 메뉴 글꼴 설정 (크기 18)
menu_font = tkfont.Font(family="Arial", size=18, weight="bold")

# 커스텀 메뉴바 영역 만들기
menu_bar_frame = tk.Frame(root, bg="lightgray", height=40)
menu_bar_frame.pack(side="top", fill="x")

# '가져오기' 메뉴 레이블 추가 (한글로 변경)
가져오기_label = tk.Label(menu_bar_frame, text="가져오기", font=menu_font, bg="lightgray", padx=20)
가져오기_label.pack(side="left")

# '필터' 메뉴 레이블 추가 (한글로 변경)
필터_label = tk.Label(menu_bar_frame, text="필터", font=menu_font, bg="lightgray", padx=20)
필터_label.pack(side="left")

# '가져오기' 메뉴 레이블 클릭 시 파일 가져오기
가져오기_label.bind("<Button-1>", lambda event: 가져오기())

# '필터' 메뉴 레이블 클릭 시 팝업 메뉴 보이기
def show_filter_menu(event):
    filter_menu.tk_popup(event.x_root, event.y_root)

# '필터' 레이블 클릭 이벤트
필터_label.bind("<Button-1>", show_filter_menu)

# 팝업 메뉴 생성 - Filter 메뉴
filter_menu = tk.Menu(root, tearoff=0, font=menu_font)
filter_menu.add_command(label="필터 기능 추가 예정")  # 필터 메뉴 항목 (추가 기능에 따라 변경 가능)

root.mainloop()
```

```python
import pandas as pd
import tkinter as tk
from tkinter import messagebox

# 데이터프레임 생성
df = voca_df

# 특정 열로 그룹을 짓고 각 그룹의 갯수 계산
group_counts = df['단어장'].value_counts().sort_index(ascending=True)

# 클릭 이벤트 핸들러
def on_item_click(item):
    messagebox.showinfo("Item Clicked", f"You clicked on {item}")

# tkinter 윈도우 생성
root = tk.Tk()
root.title("Group Count Display")

# 창의 크기를 300x700으로 설정
root.geometry("300x700")

# 리스트박스 생성 (높이와 너비 설정)
listbox = tk.Listbox(root, font=('Arial', 14), height=30, width=40)  # 폰트 크기와 리스트 크기 조절
listbox.pack(pady=20)

# 그룹화된 데이터를 리스트 형식으로 tkinter에 추가
for item, count in group_counts.items():
    display_text = f"{item} ({count})"
    listbox.insert(tk.END, display_text)

# 리스트박스 클릭 시 이벤트 처리
def on_listbox_click(event):
    # 선택된 항목 가져오기
    selection = listbox.curselection()
    if selection:
        item = listbox.get(selection[0])
        on_item_click(item)

# 리스트박스에 클릭 이벤트 바인딩
listbox.bind('<<ListboxSelect>>', on_listbox_click)

# tkinter 이벤트 루프 시작
root.mainloop()
```

```python
import pandas as pd
import tkinter as tk
from tkinter import simpledialog, messagebox

df = voca_df

# 특정 열로 그룹을 짓고 각 그룹의 갯수 계산
group_counts = df['단어장'].value_counts().sort_index()

# 인덱스(문자열) 기준으로 오름차순 정렬
sorted_group_counts = group_counts.sort_index()

# tkinter 윈도우 생성
root = tk.Tk()
root.title("Group Count Display")
root.geometry("300x700")

# 리스트박스 생성
listbox = tk.Listbox(root, font=('Arial', 14), height=30, width=40)
listbox.pack(pady=20)

# 정렬된 데이터를 리스트 형식으로 tkinter에 추가
for item, count in sorted_group_counts.items():
    display_text = f"{item} ({count})"
    listbox.insert(tk.END, display_text)

# 항목 이름 변경 함수
def change_item_name(old_item, new_name):
    # 기존 리스트에서 old_item을 찾아서 업데이트
    selection = listbox.curselection()
    if selection:
        listbox.delete(selection[0])  # 기존 항목 삭제
        listbox.insert(selection[0], new_name)  # 새로운 항목 삽입

# 리스트박스 클릭 시 이벤트 처리
def on_listbox_click(event):
    # 선택된 항목 가져오기
    selection = listbox.curselection()
    if selection:
        item = listbox.get(selection[0])
        # 항목을 분리하여 그룹 이름만 추출
        old_name = item.split(' ')[0]  # "A (3)" -> "A"
        # 새로운 이름을 입력받음
        new_name = simpledialog.askstring("Input", f"Enter new name for {old_name}:")
        if new_name:
            # 기존 갯수는 유지하고, 새로운 이름으로 항목 업데이트
            count = sorted_group_counts[old_name]  # 해당 항목의 기존 갯수
            new_item = f"{new_name} ({count})"
            change_item_name(item, new_item)

# 리스트박스에 클릭 이벤트 바인딩
listbox.bind('<<ListboxSelect>>', on_listbox_click)

# tkinter 이벤트 루프 시작
root.mainloop()
```
